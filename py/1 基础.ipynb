## 字符编码

````python
print('包含中文的str')
````

````python
print(ord('A'))
print(ord('鱼'))
print(chr(71))
print(chr(35895))
````

````python
# bytes类型的数据用带b前缀
x = b'abc'
print(x,type(x))
print('abc','abc'.encode('ascii'))
print('中文','中文'.encode('utf-8'))
# 中文编码的范围超过了ASCII编码的范围
# print('中文','中文'.encode('ascii'))

# str 换换为 byte
print('谷歌大法好'.encode('utf8'))
print(b'\xe8\xb0\xb7\xe6\xad\x8c'.decode('utf8'))
````

````python
# 部分无效的字节，使用errors='ignore'屏蔽
print(b'\xe4\xb8\xad\xff'.decode('utf-8', errors='ignore'))
# print(b'\xe4\xb8\xad\xff'.decode('utf-8'))
````

````python
# len()函数计算的是str的字符数，如果换成bytes，len()函数就计算字节数
print(len(b'ABC'))
print(len(b'\xe4\xb8\xad\xe6\x96\x87'))
print(len('中文'.encode('utf8')))
# 1个中文字符经过UTF-8编码后通常会占用3个字节，
# 而1个英文字符只占用1个字节
````

##  格式化

%s永远起作用，它会把任何数据类型转换为字符串

````python
print('谷歌%s，%s牛鼻！！\n' % ('大法好','Golang'))

print('{}苦短，我用{}！！'.format('人生','Python'))
````

````python
print('%2d-%02d' % (3, 1))
print('%.2f' % 3.1415926)
````
